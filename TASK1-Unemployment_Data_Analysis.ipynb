```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import geopandas as gpd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Dataset/Unemployment in India - Unemployment in India.csv")
data = pd.read_csv("/content/drive/MyDrive/Dataset/Unemployment_Rate_upto_11_2020 - Unemployment_Rate_upto_11_2020.csv")
```

```python
data.isnull().sum()
```

```python
data_cleaned = data.copy()
```

```python
# Convert the 'Date' column to datetime format
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'])

# Plotting Unemployment Rate over Time for different regions
plt.figure(figsize=(14, 7))
sns.lineplot(data=data_cleaned, x='Date', y='Estimated Unemployment Rate (%)', hue='Region')
plt.title('Unemployment Rate Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

```python
average_unemployment = data_cleaned.groupby('Region')['Estimated Unemployment Rate (%)'].mean().reset_index()

# Plotting the average unemployment rate by region
plt.figure(figsize=(14, 7))
sns.barplot(data=average_unemployment, x='Estimated Unemployment Rate (%)', y='Region', palette='viridis')
plt.title('Average Unemployment Rate by Region')
plt.xlabel('Average Estimated Unemployment Rate (%)')
plt.ylabel('Region')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.histplot(data_cleaned['Estimated Unemployment Rate (%)'], kde=True, color='blue', bins=30)
plt.title('Distribution of Estimated Unemployment Rate (%)')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.scatterplot(data=data_cleaned, x='Estimated Labour Participation Rate (%)', y='Estimated Unemployment Rate (%)', hue='Region', palette='viridis')
plt.title('Unemployment Rate vs. Labour Participation Rate')
plt.xlabel('Estimated Labour Participation Rate (%)')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```
